```python
import pandas as pd
```

Notebook này dùng để tiền xử lý dữ liệu. Cụ thể, chúng ta sẽ loại bỏ dữ liệu trùng (duplicate) và fill những dữ liệu có missing value.

## Dữ liệu người dùng (User)

```python
user_df = pd.read_csv('user.csv', index_col=0)
user_df
```

```python
# Remove duplicate rows
user_df.drop_duplicates(ignore_index=True, inplace=True)
user_df
```

Có vẻ như dữ liệu không xuất hiện hiện tượng trùng lặp :"D.

```python
# Figure out what columns that have missing values are
user_df.isna().sum()
```

Từ kết quả ở cell trên, ta thấy rằng chỉ mỗi `country_code` là chứa missing values, và nguyên nhân chắc chắn là do người dùng không nhập quốc gia của bản thân. Do đó ta sẽ fill các missing value này bằng chuỗi `None`.

```python
user_df.fillna(value='None', inplace=True)
user_df.isna().sum()
```

Đến đây thì dữ liệu của chúng ta đã đẹp hơn rồi :"D Lưu lại thôi nào!

```python
user_df.to_csv('user.csv')
```

## Dữ liệu Playlist
Ta cũng sẽ thực hiện tương tự các bước như đã làm ở dữ liệu người dùng.

```python
playlist_df = pd.read_csv('playlist.csv', index_col=0)
playlist_df
```

```python
# Remove duplicates rows
playlist_df.drop_duplicates(ignore_index=True, inplace=True)
playlist_df
```

Dữ liệu này cũng không xuất hiện trùng lặp :"D

```python
# Figure out what columns that have missing values are
playlist_df.isna().sum()
```

Chúng ta có hai cột chứa missing values là `genre` (thể loại nhạc) và `tracks` (id của các track có trong playlist). 
Với `genre` chúng ta sẽ fill value là `None`, còn với `tracks` chúng ta sẽ thẳng tay drop các playlist mà không có track nào :(

```python
# Drop 84 playlists that do not have any track
playlist_df.drop(labels=playlist_df[playlist_df.tracks.isna()].index, axis=0, inplace=True)
playlist_df.reset_index(inplace=True, drop=True)
playlist_df.isna().sum()
```

```python
# Fill missing values in `genre`
playlist_df.fillna('None', inplace=True)
playlist_df.isna().sum()
```

Xong, tổng kết lại chúng ta đã drop 84 dòng của dữ liệu Playlist và fill missing value bằng `None`

```python
playlist_df.to_csv('playlist.csv')
```
