## D·ªØ li·ªáu Playlist
Ta c≈©ng s·∫Ω th·ª±c hi·ªán t∆∞∆°ng t·ª± c√°c b∆∞·ªõc nh∆∞ ƒë√£ l√†m ·ªü d·ªØ li·ªáu ng∆∞·ªùi d√πng.

```python
playlist_df = pd.read_csv('playlist.csv', index_col=0)
playlist_df
```

```python
# Remove duplicates rows
playlist_df.drop_duplicates(ignore_index=True, inplace=True)
playlist_df
```

D·ªØ li·ªáu n√†y c≈©ng kh√¥ng xu·∫•t hi·ªán tr√πng l·∫∑p :"D

```python
# Figure out what columns that have missing values are
playlist_df.isna().sum()
```

Ch√∫ng ta c√≥ hai c·ªôt ch·ª©a missing values l√† `genre` (th·ªÉ lo·∫°i nh·∫°c) v√† `tracks` (id c·ªßa c√°c track c√≥ trong playlist). 
V·ªõi `genre` ch√∫ng ta s·∫Ω fill value l√† `None`, c√≤n v·ªõi `tracks` ch√∫ng ta s·∫Ω th·∫≥ng tay drop c√°c playlist m√† kh√¥ng c√≥ track n√†o :(

```python
# Drop 84 playlists that do not have any track
playlist_df.drop(labels=playlist_df[playlist_df.tracks.isna()].index, axis=0, inplace=True)
playlist_df.reset_index(inplace=True, drop=True)
playlist_df.isna().sum()
```

```python
# Fill missing values in `genre`
playlist_df.fillna('None', inplace=True)
playlist_df.isna().sum()
```

Xong, t·ªïng k·∫øt l·∫°i ch√∫ng ta ƒë√£ drop 84 d√≤ng c·ªßa d·ªØ li·ªáu Playlist v√† fill missing value b·∫±ng `None`

```python
playlist_df.to_csv('playlist.csv')
```
